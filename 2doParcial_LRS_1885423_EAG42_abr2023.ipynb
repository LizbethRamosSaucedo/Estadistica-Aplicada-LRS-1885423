1. El banco sospecha que entre mas pasa el tiempo más díficil es que puedan pagar los clientes que se les concedió un préstamo. 

a) Calcula el promedio del monto del préstamo de los que pagaron y el promedio del monto de los que no han pagado.




Dado que en la columna de credit_history se encuentran el historial crediticio el cual nos indica 
puedo poner q el evento es pagar entonces llenar con 0 y 1 


```python
pagan_prom=datos.loc[datos['full_repaid'] == 1, 'amount'].mean()
no_pagan_prom =datos.loc[datos['full_repaid'] ==0, 'amount'].mean()
```

```python
#imprimir 
print(f' \nEl promedio del monto del préstamo de los que pagaron es de  {round(pagan_prom,2)} \n El promedio del monto del préstamo de los que no pagaron es de {round(no_pagan_prom,2)}')
```

b) Ajusta un modelo weibull y un modelo log-normal y graficas las funciones de riesgo respectivas

```python
from lifelines import WeibullFitter 
from lifelines import LogNormalFitter 
```

```python
weib = WeibullFitter().fit(datos['duration'],datos['full_repaid'])
weib.plot_hazard()
plt.title(' Función de riesgo Weibull ')
plt.ylabel('Función de riesgo')
plt.xlabel ('Tiempo en meses')
plt.show();
```

```python
logn = LogNormalFitter().fit(datos['duration'],datos['full_repaid'])
logn.plot_hazard()
plt.title(' Función de riesgo Log-normal ')
plt.ylabel('Función de riesgo')
plt.xlabel ('Tiempo en meses')
plt.show();
```

c) comenta que significa la grafica de la función de riesgo de la weibull respecto a si los clientes tardan en pagar


La función de riesgo mediante el estimador Weibull es creciente conforme al tiempo lo cual significa que el incumplimiento del pago aumenta conforme pasa el tiempo.

d) comenta que significa la grafica de la función de riesgo de la log-normal respecto a si los clientes tardan en pagar

 Log-Normal se caracteriza por tener una tasa de crecimiento exponencial al principio y luego estabilizarse a medida que transcurre el tiempo.
 
 
En la grafica se puede observar que al principio la distibución log normal es creciente aproximadamente hasta el mes 20 y despues se mantiene y  decrece por lo cual significa que el mayor riesgo de incumplimiento de pago se da entre el tiempo 20-30 meses y apartir de ahi el riesgo disminuye.

2. Estima la función de supervivencia por el método de Kaplan-Meier para el tiempo que tardan los clientes en pagar y muestra su gráfica correspondiente

```python
from lifelines import KaplanMeierFitter
```

```python
kt = KaplanMeierFitter().fit(datos['duration'],datos['full_repaid'])
kt.plot_survival_function(label=' tiempo que tardan los clientes en pagar',ci_show=False, color='orange')
plt.title(' Función de supervivencia de Kaplan-Meier ')
plt.ylabel('Probabilidad de supervivencia')
plt.xlabel ('Tiempo en meses')
plt.show();
```

3. Coloca la gráfica de kaplan Meier sobre la grafica de supervivencia log-normal 

```python
logn.plot_survival_function( label='Supervivencia log-normal') 
kt.plot_survival_function( label='Supervivencia Kaplan-Meier ')
plt.title(' Función de supervivencia de Kaplan-Meier y log-normal ')
plt.ylabel('Probabilidad de supervivencia')
plt.xlabel ('Tiempo en meses')
plt.show();

```

4. Comenta si las graficas se parecen o no, y en que se diferencian

Tanto la función de supervivencia de Kaplan-Meier y log-normal se muestran de manera decreciente con el tiempo, entonces lo que se visualiza en la gráfica es que probabilidad de que el prestatario pague su préstamo disminuye con el tiempo.
La diferencia entre estos dos modelos es que Kapla-Meier es un modelo de supervivencia no paramétrico y gráficamente se muestra en forma de escalera, ahi se visualiza los prestatarios que sobreviven en distintos tiempos.
La gráfica de supervivencia log-normal, es un modelo paramétrico y muestra una curva  que representa la probabilidad de supervivencia a diferentes tiempos.

$H_a:\beta_i \neq  0$ Variable si afecta

$\alpha = 0.05$

Si $p>\alpha$ entonces se rechaza $H_o$

Si coef<0 : el riesgo disminuye en 1-exp(coef) = %

Si coef>0 : el riesgo aumenta en exp(coef)-1 = %
