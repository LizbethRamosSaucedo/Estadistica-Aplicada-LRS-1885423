```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[1,8,27,64, 125]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, color='red')
```

```python
#data frame a partir de un diccionario 
mun_nl=pd.DataFrame({'municipio':['Monterrey','Guadalupe','Apodaca', 'San Nicolas'],'nacimientos':[500,630,768,400]})
mun_nl
```

```python
plt.bar(mun_nl['municipio'], mun_nl['nacimientos'], color=['blue','green','red','yellow'])
       #agregar titulo
plt.title('Nacimientos por municipio')
        #agregar nombre al eje x 
plt.xlabel('Municipios de NL ')
        #Agregar nombre a eje y 
plt.ylabel('Cantidad de nacimientos')
        #cambiar los saltos escala y 
plt.yticks([0,200,400,600,800])
        #mostrar grafuca 
plt.show()
```

```python
mun_nl =mun_nl.sort_values('nacimientos',ascending= False)
plt.bar(mun_nl['municipio'], mun_nl['nacimientos'], color=['blue','green','red','yellow'])
       #agregar titulo
plt.title('Nacimientos por municipio')
        #agregar nombre al eje x 
plt.xlabel('Municipios de NL ')
        #Agregar nombre a eje y 
plt.ylabel('Cantidad de nacimientos')
        #cambiar los saltos escala y 
plt.yticks([0,200,400,600,800])
        #mostrar grafuca 
plt.show()
```

```python
fig,(panel1, panel2) = plt.subplots(2,1) #subgraficas, su nombre 
fig.tight_layout()
#panel 1 (subgrafica 1)
x1=x #eje x de grafica 1
y1= y #eje y de grafica 1
panel1.plot(x,y, label='nombre de linea')
panel1.title.set_text('Grafica 1')

#panel 2 (subgrafica 2 )
x2=mun_nl['municipio']
y2=mun_nl['nacimientos']
panel2.bar(x2,y2)
panel2.title.set_text('Grafica 2')

plt.show()
```
