```python
#las de siempre 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import adfuller #prueba de dickey-fuller
```

```python
#datos 
procesos= pd.read_excel('procesos_estacionariosG41.xlsx')
```

```python
procesos.head()
```

Primero analizaremos el ruido blanco (white noise/proceso aleatorio puro)

```python
procesos['ruido_blanco'].plot(color='black', linewidth=0.3)
plt.xlim(0,1000)
plt.xlabel('tiempo')
plt.title('Ruido blanco')
plt.savefig('ruido_blanco.jpg')
```

1. Determinar si la serie (ruido blanco) es estacionario con una prueba de ra√≠z unitaria 

```python
dickeyfuller1 = adfuller(procesos['ruido_blanco'])
alpha = 0.05

if dickeyfuller1[1]<alpha:
 print('Se rechaza H0, la serie es estacionaria ')
else: 
    print('No se rechaza H0, la serie es no estacionaria')

#p_value > alpha rechazamos H0
```

2. Determinar que tipo de proceso estacionario es con las gr√°ficas de autocorrelaci√≥n ACF y autocorrelaci√≥n parcial 

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
#subgrafica 0 
plot_acf(procesos['ruido_blanco'], zero=False, ax=ax[0], title= 'Funci√≥n de autocorrelaci√≥n ACF')
ax[0].set_ylim(-1,1)
#subgrafica 1 
plot_pacf(procesos['ruido_blanco'], zero=False,ax=ax[1], title= 'Funci√≥n de autocorrelaci√≥n parcial PACF' )
ax[1].set_ylim(-1,1)
plt.savefig('correl_ruidoblanco.jpg')
```
