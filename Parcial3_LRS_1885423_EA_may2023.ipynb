Arthur Burns y Wesley Michell en 1946 en su libro "Measuring Bussiness Cycles" definen al ciclo económico como:

"…*las fluctuaciones en la actividad agregada de un país que se encuentra organizada en empresas de negocios*…"

**El ciclo económico clásico se caracteriza por tres fases principales:** 

I.	**Recesión**: es la etapa donde la actividad económica comienza a disminuir y termina cuando la actividad económica alcanza su punto más bajo. 

II.	**Recuperación**: comienza cuando la actividad económica comienza a crecer despues de una recesión y termina cuando la actividad alcanza el nivel del pico donde comenzó la recesión.

III. **Expansión**: Es cuando la actividad económica se está expandiendo y supera el nivel del pico de la recesión.

```python
#Librerias para pruebas 
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
```

```python
#1.PRUEBA DE RAIZ UNITARIA 

ADF_test1=adfuller(pib_trim['PIB'])

#Ho: Serie No Estacionaria
#Ha: Serie Estacionaria

alpha=0.05
pvalue=ADF_test1[1]
if pvalue<alpha:
  print('Rechazamos Ho, Serie Estacionaria')
else:
  print('No rechazamos Ho, Serie No Estacionaria')
```

```python
#2.En caso de que no lo sea diferencia el PIB y guardalo en un objeto llamado pib_diff
pib_trim['PIB_diferenciado'] = pib_trim['PIB'].diff()
```

```python
pib_trim
```

```python
#3.Realiza la prueba de raíz unitaria y verifica que la primera diferencia del PIB sea estacionario

ADF_test2=adfuller(pib_trim['PIB_diferenciado'].dropna())

#Ho: Serie No Estacionaria
#Ha: Serie Estacionaria

alpha=0.05
pvalue=ADF_test2[1]
if pvalue<alpha:
  print('Rechazamos Ho, Serie Estacionaria')
else:
  print('No rechazamos Ho, Serie No Estacionaria')
```

```python
#4.Una vez que el PIB sea estacionario realiza las gráficas de ACF y PACF para tratar de identificar el proceso, 
#y plantea por lo menos tres procesos (solo plantearlos, no estimarlos)

ig, ax = plt.subplots(1,2,figsize=(12,4))

plot_acf(pib_trim['PIB_diferenciado'].dropna(), zero=False, lags=40, ax=ax[0],title="Función de Autocorrelación")
ax[0].set_ylim(-1,1)
plot_pacf(pib_trim['PIB_diferenciado'].dropna(), zero=False, lags=40, ax=ax[1], title="Función de Autocorrelación Parcial")
ax[1].set_ylim(-1,1)
plt.show()
```

ACF te da el grado del MA(q)

PACF te da el grado del AR(p)


Un modelo ARIMA(p,I,q) es un AR de orden p, con MA de orden q y con un orden determinado de integracion I. El orden de integracion es cuantas veces se le aplico primeras diferencias a la serie para volverla estacionaria. 

Se plantean los siguientes modelos: 
AR(2),
ARMA(2,1), 
MA(1),
AR(1) y 
ARMA(1,1) 


```python
#5.Con base en la notebook "pronostico_arima" utiliza la función "for" para probar hasta un arma(6,6) de la series pib_diff
#vamos a estimar varios modelos y que se guarde el AIC BIC
ic = {}
for ar in range(7):
    for ma in range(7):
        print(f"AR: {ar}, MA: {ma}")
        modelo = ARIMA(pib_trim['PIB'].dropna(), order=(ar,1, ma))
        resultado = modelo.fit()
        ic[(ar, ma)] = [resultado.aic, resultado.bic]

 

ic = pd.DataFrame(ic, index=["AIC", "BIC"]).T
ic.index = ic.index.set_names(["AR", "MA"])
```

```python
ic.sort_values('AIC', ascending=True).head(3)
```
